```python
#Example 1
import unicodedata
import pandas as pd


# ==============================
# 🔹 1. FUNCION PARA LIMPIAR TEXTO
# ==============================
def limpiar(texto):
    if pd.isna(texto):
        return ""
    texto = str(texto).lower()
    texto = unicodedata.normalize('NFKD', texto).encode('ascii', 'ignore').decode('utf-8')
    return texto

# 🔹 Tus categorías (puedes ampliarlas)
categorias_T = {
    "Alcaldías": ["alcaldia"],
    "Gobernaciones": ["gobernacion"],
    "ARN": ["arn"],
    "Defensoría del Pueblo": ["defensoria"],
    "Procuraduría General de la Nación": ["procuraduria"],
    "UBPD": ["unidad de busqueda de personas dadas por desaparecidas"],
    "UARIV": ["unidad para la atencion"],
    "ART": ["agencia de renovacion"],
    "URT": ["restitucion de tierras"],
    "ANT": ["agencia nacional de tierras"],
    "Mision Verificacion de las Naciones Unidas": ["mision"],
    "GIZ": ["giz"],
    "Personerias": ["personeria"],
    "Víctimas no colectivizadas": ["victimas no colectivizadas"],
    "Organización de Mujeres": ["mujeres"],
    "Organización LGBTIQ+": ["lgbtiq"]
}

categorias_S= {
    "Universidades": ["academia"]
}

# 🔹 Función clave
def clasificar(fila):
    resultado = set()

    # 🔸 Evaluar columna T (ej: entidades)
    texto_T = limpiar(fila["Entidad"])  # 👈 cambia si tu columna se llama distinto
    partes = texto_T.split(",")

    for parte in partes:
        for categoria, palabras in categorias_T.items():
            for palabra in palabras:
                if palabra in parte:
                    resultado.add(categoria)
                    break

    # 🔸 Evaluar columna S (categoria especial)
    texto_S = limpiar(fila["Tipo de organización"])  # 👈 cambia por el nombre real
    for categoria, palabras in categorias_S.items():
        for palabra in palabras:
            if palabra in texto_S:
                resultado.add(categoria)
                break

    return ", ".join(sorted(resultado))

# 🔹 Aplicación
df = pd.read_excel(r"ruta_archivo.xlsx", sheet_name="Base_limpia")

df["Categorias"] = df.apply(clasificar, axis=1)

df.to_excel("resultado.xlsx", index=False)

print("✅ Proceso terminado. Archivo generado: resultado.xlsx")
```
